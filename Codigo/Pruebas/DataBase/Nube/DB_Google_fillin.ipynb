```python
import mysql.connector
import re

def es_matricula_valida_alumno(matricula):
    return re.match(r'^A\d{8}$', matricula) is not None

def es_matricula_valida_profesor(matricula):
    return re.match(r'^L\d{8}$', matricula) is not None

# Configura las credenciales de la base de datos
config = {
    'user': 'root',
    'password': 'Jorge_1419',
    'host': '34.121.148.111',
    'database': 'proyecto'
}

# Conecta a la base de datos
con = mysql.connector.connect(**config)
cursor = con.cursor()

while True:
    print("\n1. Alumno")
    print("2. Profesor")
    print("3. Salir")
    seleccion = input("Seleccione una opción (1/2/3): ")
    
    if seleccion == "3":
        break
    elif seleccion == "1":
        tipo_persona = "alumno"
    elif seleccion == "2":
        tipo_persona = "profesor"
    else:
        print("Opción no válida.")
        continue

    matricula = input("Ingrese la matrícula: ")

    if tipo_persona == "alumno" and not es_matricula_valida_alumno(matricula):
        print("Matrícula de alumno inválida. Debe ser del formato A01749489.")
        continue
    elif tipo_persona == "profesor" and not es_matricula_valida_profesor(matricula):
        print("Matrícula de profesor inválida. Debe ser del formato L01749489.")
        continue

    # Insertamos en la tabla personas primero
    cursor.execute("INSERT INTO personas (matricula, tipo) VALUES (%s, %s)", (matricula, tipo_persona))
    con.commit()

    if tipo_persona == "alumno":
        nombre = input("Ingrese el nombre del alumno: ")
        apellido = input("Ingrese el apellido del alumno: ")
        carrera = input("Ingrese la carrera del alumno: ")
        hora_inicio = input("Ingrese la hora de inicio (formato HH:MM:SS): ")
        hora_fin = input("Ingrese la hora de fin (formato HH:MM:SS): ")

        # Inserta en la tabla alumnos
        cursor.execute("INSERT INTO alumnos (matricula1, nombre, apellido, carrera, hora_de_inicio, hora_de_FIN) VALUES (%s, %s, %s, %s, %s, %s)", (matricula, nombre, apellido, carrera, hora_inicio, hora_fin))
        con.commit()

        # Registrar participación
        fecha_participacion = input("Ingrese la fecha de participación (formato YYYY-MM-DD): ")
        comentario = input("Ingrese un comentario sobre la participación: ")
        cursor.execute("INSERT INTO participacion (fecha, comentario, alumno_id) VALUES (%s, %s, %s)", (fecha_participacion, comentario, matricula))
        con.commit()

    elif tipo_persona == "profesor":
        nombre = input("Ingrese el nombre del profesor: ")
        apellido = input("Ingrese el apellido del profesor: ")
        edad = input("Ingrese la edad del profesor: ")

        # Inserta en la tabla profesores
        cursor.execute("INSERT INTO profesores (matricula2, nombre, apellido, edad) VALUES (%s, %s, %s, %s)", (matricula, nombre, apellido, edad))
        con.commit()

    # Registrar asistencia (común para ambos tipos)
    fecha_asistencia = input("Ingrese la fecha de asistencia (formato YYYY-MM-DD): ")
    asistio = input(f"¿Asistió el {tipo_persona}? (Sí=1, No=0): ")
    cursor.execute("INSERT INTO asistencia (fecha, matricula, asistio, tipo) VALUES (%s, %s, %s, %s)", (fecha_asistencia, matricula, asistio, tipo_persona))
    con.commit()

# Cierra la conexión
cursor.close()
con.close()

print("\n¡Gracias por usar el sistema!")
```
